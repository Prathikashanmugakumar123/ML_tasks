```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
import pandas as pd
data=pd.read_csv("Salary_dataset.csv")
data
```

```python
exp=data["YearsExperience"]
exp
```

```python
salary=data["Salary"]
salary
```

```python
model=LinearRegression()
```

```python
x=np.array(exp).reshape(-1,1)
y=np.array(salary).reshape(-1,1)
```

```python
x
```

```python
y
```

```python
model.fit(x,y)
```

```python
new_expirence=int(input("Enter your experience:"))
new_data=np.array(new_expirence).reshape(-1,1)
y_answer=model.predict(new_data)
print("Your salary is:",model.predict(new_data))
```

```python
plt.plot(x,y)
plt.scatter(new_data,y_answer)
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Salary vs Experience")
plt.show()
```
