```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder

watch_time = ["High", "Low", "Medium", "High", "Medium", "Low"]
genre_match = ["Yes", "Yes", "No", "Yes", "Yes", "No"]
weekend = ["Yes", "No", "Yes", "No", "Yes", "No"]

like_movie = ["Like", "Not Like", "Not Like", "Like", "Like", "Not Like"]

le_watch = LabelEncoder()
le_genre = LabelEncoder()
le_weekend = LabelEncoder()
le_target = LabelEncoder()

watch_time_enc = le_watch.fit_transform(watch_time)
genre_match_enc = le_genre.fit_transform(genre_match)
weekend_enc = le_weekend.fit_transform(weekend)
like_movie_enc = le_target.fit_transform(like_movie)


X = list(zip(watch_time_enc, genre_match_enc, weekend_enc))
y = like_movie_enc

print("Encoded Features (X):", X)
print("Encoded Target (y):", y)

model = RandomForestClassifier(
    n_estimators=3,  
    random_state=42
)

model.fit(X, y)

new_user = [[
    le_watch.transform(["Medium"])[0],
    le_genre.transform(["Yes"])[0],
    le_weekend.transform(["No"])[0]
]]

prediction = model.predict(new_user)
result = le_target.inverse_transform(prediction)


print("\nNew User Prediction:")
print("Watch Time : Medium")
print("Genre Match: Yes")
print("Weekend    : No")
print("Result     :", result[0])
```

```python
# Random Forest Regression 

from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
weather = ["Hot", "Cold", "Warm", "Hot", "Warm"]
promotion = ["Yes", "No", "Yes", "No", "No"]

sales = [80, 30, 60, 50, 40]

le_weather = LabelEncoder()
le_promo = LabelEncoder()

weather_enc = le_weather.fit_transform(weather)
promo_enc = le_promo.fit_transform(promotion)

X = list(zip(weather_enc, promo_enc))
y = sales

print("Encoded Features:", X)
print("Target:", y)
model = RandomForestRegressor(
    n_estimators=3,    
    random_state=42
)

model.fit(X, y)

new_data = [[
    le_weather.transform(["Warm"])[0],
    le_promo.transform(["Yes"])[0]
]]

predicted_sales = model.predict(new_data)
print("Predicted Sales for New Day:", predicted_sales[0])

```
