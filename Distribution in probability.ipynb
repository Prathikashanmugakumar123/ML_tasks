```python
from scipy.stats import norm
# Given values
mu = 170  # mean
sigma = 10  # standard deviation
X = 180  # value to check
prob_greater = 1 - norm.cdf(X, mu, sigma)
print("Probability student is taller than 180 cm:", round(prob_greater * 100, 2), "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Normal Distribution parameters
mu = 170
sigma = 10

# Create random sample dataset
X = np.random.normal(mu, sigma, 10000)

# Histogram (Bell Shape)
plt.hist(X, bins=30, density=True, alpha=0.6)

# Plot normal distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 1000)
y = norm.pdf(x, mu, sigma)
plt.plot(x, y, linewidth=2)

# Labels and title
plt.title("Normal Distribution Histogram\nBell Shaped Curve")
plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import expon
import numpy as np
import matplotlib.pyplot as plt

# Given values
mean_time = 10  
t = 5  

lambda_value = 1 / mean_time
prob_within = expon.cdf(t, scale=1/lambda_value)

print("Probability customer arrives within 5 minutes:",
      round(prob_within * 100, 2), "%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

mean_time = 10  # average customer arrival time

# generate random exponential data
data = np.random.exponential(mean_time, 10000)

# histogram
plt.hist(data, bins=30, density=True)  

plt.title("Exponential Distribution - Histogram")
plt.xlabel("Time (minutes)")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from scipy.stats import uniform
import numpy as np
import matplotlib.pyplot as plt
# Uniform Distribution range
a = 1      # minimum
b = 100    # maximum
# Interval to check
low = 20
high = 40
dist = uniform(loc=a, scale=b-a)
probability = dist.cdf(high) - dist.cdf(low)

print("Probability of getting number between 20 and 40:",
      round(probability * 100, 2), "%")
```
