```python
# classification in random forest
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier as rc
data={"hrs":[2,5,1,7,3,6],
     "attendence":[50,70,40,90,60,80],
     "pass":[0,1,0,1,0,1]}
f=pd.DataFrame(data)
x=f[["hrs","attendence"]]
y=f[["pass"]]
randomfor=rc(n_estimators=5,random_state=0)
randomfor.fit(x,y)
new_std=[[4,65]]
predition=randomfor.predict(new_std)
print(predition)
```

```python
#regression in random forest
from sklearn.ensemble import RandomForestRegressor as rf
data={
    "size":[1200,1500,800,1800,1100],
    "Bedrooms":[2,3,1,3,2],
    "price":[200,250,150,300,180]
}
f1=pd.DataFrame(data)
x=f1[["size","Bedrooms"]]
y=f1[["price"]]
new_house=[[1400,2]]
regressionss=rf(n_estimators=5,random_state=0)
regressionss.fit(x,y)
predicition=regressionss.predict(new_house)
print(predicition)
```
