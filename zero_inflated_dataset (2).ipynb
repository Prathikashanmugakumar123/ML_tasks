```python
import pandas as pd
import statsmodels.api as sm

from statsmodels.discrete.count_model import ZeroInflatedPoisson
```

```python
df = pd.read_csv("insurance.csv")
df
```

```python
X = df[["age", "bmi"]]
X
```

```python
y = df["children"]
y
```

```python
X = sm.add_constant(X)
X
```

```python
model = ZeroInflatedPoisson(
    endog=y,
    exog=X,
    exog_infl=X
)

result = model.fit()
```

```python
print(result.summary())
```

```python
print(df["children"].value_counts())
```

```python
zero_percentage = (df["children"] == 0).mean() * 100

print("Zero Percentage:", round(zero_percentage,2), "%")
```

```python
age = int(input("Enter Age: "))
bmi = float(input("Enter BMI: "))

new_data = pd.DataFrame({
    "const": [1],
    "age": [age],
    "bmi": [bmi]
})

prediction = result.predict(
    exog=new_data,
    exog_infl=new_data
)

print("Predicted Children:", round(prediction[0]))
```
