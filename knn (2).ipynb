```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# Dataset
# Weight (g), Texture (0 = Smooth, 1 = Rough)
X = np.array([
    [150, 0],  # Apple
    [170, 0],  # Apple
    [140, 1],  # Orange
    [130, 1]   # Orange
])

# Labels
y = np.array(["Apple", "Apple", "Orange", "Orange"])

# Create KNN model (k = 3)
model = KNeighborsClassifier(n_neighbors=3)

# Train model
model.fit(X, y)

# Predict new fruit
new_fruit = [[160, 0]]   # Weight=160g, Texture=Smooth
prediction = model.predict(new_fruit)

print("Predicted Fruit:", prediction[0])
```
