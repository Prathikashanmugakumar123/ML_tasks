```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.preprocessing import LabelEncoder

# Data
emails = [
    "win money",
    "free prize",
    "hello friend",
    "how are you"
]

labels = ["Spam", "Spam", "Not Spam", "Not Spam"]

# Encode output labels
le = LabelEncoder()
y = le.fit_transform(labels)

# Encode input text
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Train model
model = MultinomialNB()
model.fit(X, y)

# Test
test_email = ["win free"]
test_vec = vectorizer.transform(test_email)

pred = model.predict(test_vec)

# Decode result
print("Prediction:", le.inverse_transform(pred)[0])
```

```python
# Import libraries
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder

X = np.array([[40], [45], [50], [60], [70]])

y_text = ["Fail", "Fail", "Fail", "Pass", "Pass"]

# Encode labels (Fail=0, Pass=1)
le = LabelEncoder()
y = le.fit_transform(y_text)

model = GaussianNB()


model.fit(X, y)

new_marks = [[55]]
prediction = model.predict(new_marks)

# Decode result
print("Predicted Result for 55 marks:", le.inverse_transform(prediction)[0])
```

```python

import numpy as np
from sklearn.naive_bayes import BernoulliNB
from sklearn.preprocessing import LabelEncoder

# Input features (Spicy, Cheese)
X = np.array([
    [1, 1],
    [1, 0],
    [0, 1],
    [0, 0]
])

y_text = ["Yes", "Yes", "No", "No"]

le = LabelEncoder()
y = le.fit_transform(y_text)

model = BernoulliNB()

model.fit(X, y)

# Predict for new dish: Spicy=1, Cheese=0
new_dish = [[1, 0]]
prediction = model.predict(new_dish)

# Decode prediction
print("Prediction:", le.inverse_transform(prediction)[0])
```
